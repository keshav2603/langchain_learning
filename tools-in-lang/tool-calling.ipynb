```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from dotenv import load_dotenv
load_dotenv()

llm=HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-chat")

model=ChatHuggingFace(llm=llm,)
```

```python
model.invoke("hello")
```

```python
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
@tool
def multiply(a:int,b:int)->int:
    """given 2 number a,b this tool will return thier product """
    return a*b
```

```python
#tool binding
```

```python
model_tools=model.bind_tools([multiply])
```

```python
model.invoke("product of 3 and 7")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)
```

```python
query=HumanMessage("what is the product of 3 and 6")
message=[query]
```

```python
llm_tools=llm.bind_tools([multiply])
```

```python
result=llm_tools.invoke(message)
```

```python
message.append(result)
message
```

```python
tools_result=multiply.invoke(result.tool_calls[0])
```

```python
message.append(tools_result)
message
```
