```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEmbeddings

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_chroma import Chroma

from langchain_core.prompts import PromptTemplate,SystemMessagePromptTemplate

from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)
```

```python
llm.invoke("hello")
```

```python
loader = PyPDFLoader("sample.pdf")
```

```python
docs=loader.load()
```

```python
docs[0]
```

```python
splitter= RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
```

```python
chunks=splitter.create_documents([text.page_content for text in docs])
```

```python
chunks
```

```python
len(chunks)
```

```python
embedding_model =HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_store=Chroma(
    embedding_function=embedding_model,
    collection_name="sampll1"
)
```

```python
vector_store.add_documents(chunks)
```

```python
retriever=vector_store.as_retriever(search_type="similarity",search_kwargs={"k":5})
```

```python

prompt = PromptTemplate(template="""
You are an expert AI Research Assistant specialized in explaining academic research papers clearly and accurately.

You must answer ONLY using the provided CONTEXT.

======================
CONTEXT:
{context}
======================

QUESTION:
{question}

======================

STRICT RULES:

1. Use ONLY the information in the CONTEXT.
2. If the answer is not clearly supported by the context, say:
   "The provided context does not contain enough information to answer this."
3. Do NOT make assumptions.
6. Keep explanations structured and clear.

RESPONSE FORMAT:

1. Short Summary  
2. Detailed Explanation  
3. Key Points (bullet list if helpful)  
4. Citations  

Be precise, grounded, and academically reliable.
""",input_variables=["context","question"])
```

```python
question="type of layeers in cnn"
context=retriever.invoke(question)
```

```python
context
```

```python
final_prt=prompt.invoke({"question":question,"context":context})
```

```python
llm.invoke(final_prt).text
```
