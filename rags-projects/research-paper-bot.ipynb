```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEmbeddings

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_chroma import Chroma

from langchain_core.prompts import PromptTemplate,SystemMessagePromptTemplate

from langchain_core.messages import AIMessage,HumanMessage

from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)
```

```python
llm.invoke("hello")
```

```python
loader = PyPDFLoader("sample.pdf")
```

```python
docs=loader.load()
```

```python
docs[0]
```

```python
splitter= RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
```

```python
chunks=splitter.create_documents([text.page_content for text in docs])
```

```python
chunks
```

```python
len(chunks)
```

```python
embedding_model =HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_store=Chroma(
    embedding_function=embedding_model,
    collection_name="sampll1"
)
```

```python
vector_store.add_documents(chunks)
```

```python
retriever=vector_store.as_retriever(search_type="similarity",search_kwargs={"k":5})
```

```python

prompt = PromptTemplate(template="""
You are an expert AI Research Assistant specialized in explaining academic research papers.
Answer ONLY using the provided CONTEXT and CHAT HISTORY.

======================
CHAT HISTORY:
{chat_history}
======================
CONTEXT:
{context}
======================
QUESTION: {question}
======================

STRICT RULES:
1. Use ONLY the information in the CONTEXT.
2. Take CHAT HISTORY into account for follow-up questions.
3. If the answer is not in the context, say: "The provided context does not contain enough information to answer this."
4. Keep explanations structured and clear.

RESPONSE FORMAT:
1. Short Summary
2. Detailed Explanation
3. Key Points (bullet list if helpful)
4. Citations
""", input_variables=["context", "question", "chat_history"])
```

```python
question="type of layeers in cnn"
context=retriever.invoke(question)
```

```python
final_prt=prompt.invoke({"question":question,"context":context})
```

```python
def format_history(chat_history):
    """Convert list of messages into a readable string (last 10 only)."""
    
    if not chat_history:
        return "No previous conversation."
    
    # ✅ Take only last 10 messages
    recent_history = chat_history[-10:]
    
    lines = []
    for msg in recent_history:
        if isinstance(msg, HumanMessage):
            lines.append(f"User: {msg.content}")
        elif isinstance(msg, AIMessage):
            lines.append(f"Assistant: {msg.content}")
    
    return "\n".join(lines)
```

```python
def chat(question,chat_history):

    context=retriever.invoke(question)
    history_str=format_history(chat_history)
    final_prompt = prompt.invoke({
        "question": question,
        "context": context,
        "chat_history": history_str
    })
    response=llm.invoke(final_prompt).content

    chat_history.append(HumanMessage(content=question))
    chat_history.append(AIMessage(content=response))

    return response, chat_history
```

```python
def main():
    print("📄 PDF Research Assistant (type 'exit' to quit)\n")
    chat_history = []
    
    while True:
        question = input("You: ").strip()
        
        if not question:
            continue
        if question.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break
        
        response, chat_history = chat(question, chat_history)
        print(f"\nAssistant: {response[0].text}\n")
        print("-" * 60)

main()
```
